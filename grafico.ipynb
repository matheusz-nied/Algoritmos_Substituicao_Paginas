```python
import matplotlib.pyplot as plt
import pandas as pd

arquivo_csv = 'resultados.csv'
dados = pd.read_csv(arquivo_csv)

quadros = dados['Molduras']
fifo = dados['FIFO']
segunda_chance = dados['Segunda Chance']
relogio = dados['Relgio']

plt.figure(figsize=(10, 6))
plt.plot(quadros, fifo, marker='o', label='FIFO')
plt.plot(quadros, segunda_chance, marker='s', label='Segunda Chance')
plt.plot(quadros, relogio, marker='^', label='Relgio')

plt.title('Desempenho dos Algoritmos de Substituio de Pginas')
plt.xlabel('Nmero de Molduras (Quadros)')
plt.ylabel('Nmero Mdio de Faltas de Pgina')
plt.legend()
plt.grid(True)


plt.show()

```
